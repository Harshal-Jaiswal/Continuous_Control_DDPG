For this project, we are using the __Deep Deterministic Policy Gradient (DDPG) Algorithm__.
    with the following hyperparameter

    BUFFER_SIZE = int(1e6)  
    BATCH_SIZE = 1024       (minibatch size)
    GAMMA = 0.99            (discount factor)
    TAU = 1e-3              (soft update parameters)
    LR_ACTOR = 1e-4         (actor learning rate)
    LR_CRITIC = 3e-4        (critic learning rate) 
    WEIGHT_DECAY = 0.0001   (L2 weight decay)
    LEAKINESS = 0.01        (LeakyReLU )

  The 2 Models Actor and Critic uses the following architecture
     - Actor
       It uses 2 layers with 256 and 128 units and a leaky-Relu activation function and a tanh activation function for the output layer.
      - Critic
       It uses 3 layers with 256, 256+ action state size, 128 units, and l leaky-Relu activation function and no activation function in the output layer.
       
  following techniques were also implemented
           - Soft Updates
           - Experience Replay
           - Fixed targets

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

DDPG_scores = np.load('./ddpg_scores.npy')

f = plt.figure()
ax = f.add_subplot(111)
plt.plot(range(1, len(DDPG_scores) + 1), DDPG_scores)
plt.xlabel('# Episodes')
plt.ylabel('Scores')
plt.savefig('scores_plot.png')
plt.show()
```

The Agent reached the goal score of 30 at 183 episodes. With 20 agents the score gained quickly and reached its desired position.
